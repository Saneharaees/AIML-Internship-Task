## 📝 Task 2 — Grouped Bar Chart with Matplotlib

### Background
A **grouped bar chart** places multiple bars side by side for the same category, making it easy to compare values within each group.

### 🎯 Your Task
The data below shows **exam scores** for three subjects across two classes.

| Subject | Class A | Class B |
|---------|---------|---------|
| Math | 72 | 68 |
| Science | 85 | 80 |
| English | 78 | 74 |
| History | 65 | 70 |

Create a **grouped bar chart** with:
- Two bars (Class A and Class B) side by side for each subject
- Different colors for each class
- Value labels on top of each bar
- A legend, axis labels, and a title

> 💡 **Hint:** Use `np.arange(len(subjects))` for x-positions, then offset each group with `x - 0.2` and `x + 0.2` and `width=0.4`.  
> To add labels: loop over bars using `bar.get_height()` and `plt.text()`.


````python
import numpy as np

subjects = ['Math', 'Science', 'English', 'History']
class_a  = [72, 85, 78, 65]
class_b  = [68, 80, 74, 70]

# YOUR CODE HERE
# Create x positions using np.arange
x = np.arange(len(subjects))


# Plot Class A bars (offset left)
bars=plt.bar(x - 0.2, class_a, width=0.4, label='Class A', color='blue')




# Plot Class B bars (offset right)
bars1=plt.bar(x + 0.2, class_b, width=0.4, label='Class B', color='green')


# Add value labels above each bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, f'{height}', ha='center', va='bottom')


for bar1 in bars1:
    height = bar1.get_height()
    plt.text(bar1.get_x() + bar1.get_width() / 2, height, f'{height}', ha='center', va='bottom')



# Add title, axis labels, legend, and x-tick labels
plt.title('Exam Scores by Subject and Class')
plt.xlabel('Subjects')
plt.ylabel('Scores')
plt.xticks(x, subjects)
plt.legend()
plt.show()

````

### 🎯 Your Task
Load the `tips` dataset and create the following **3 plots** (each in its own cell or in subplots):

**Plot 1 — Box Plot:** Distribution of `tip` amounts grouped by `day`, with `sex` as the hue.

## 📝 Task 7 — Interactive Line Plot with Bokeh

### Background
Bokeh creates **interactive, browser-rendered** plots. The `HoverTool` lets users see exact data values when they hover over points.

### 🎯 Your Task
Create an **interactive Bokeh line plot** showing website traffic over 7 days:

```
days    = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
visitors = [120, 180, 150, 200, 250, 300, 270]
```

Requirements:
- Line width = 2, color of your choice
- Circle markers on each data point (size = 8)
- A `HoverTool` that shows the day and visitor count on hover
- Proper title and axis labels

> 💡 **Hint:** To use HoverTool with plain lists, convert data to a `ColumnDataSource`.  
> `source = ColumnDataSource(data=dict(days=days, visitors=visitors))`  
> Then use `p.line(x='days', y='visitors', source=source)` and  
> `HoverTool(tooltips=[("Day", "@days"), ("Visitors", "@visitors")])`.


````python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import HoverTool, ColumnDataSource

output_notebook()

days     = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
visitors = [120, 180, 150, 200, 250, 300, 270]

# YOUR CODE HERE
# 1. Create a ColumnDataSource
source = ColumnDataSource(data=dict(days=days, visitors=visitors))

# 2. Create a Bokeh figure with title and axis labels
p = figure(x_range=days, height=350, title="Website Traffic Over 7 Days",
           x_axis_label="Day", y_axis_label="Visitors")

# 3. Add a line renderer
p.line(x='days', y='visitors', source=source, line_width=2, color='blue', legend_label='Visitors')

# 4. Add circle markers (changed from p.circle to p.scatter to avoid deprecation warning)
p.scatter(x='days', y='visitors', source=source, size=8, color='blue')

# 5. Add a HoverTool
hover = HoverTool(tooltips=[("Day", "@days"), ("Visitors", "@visitors")])
p.add_tools(hover)

# 6. Show the plot
show(p)
````

Use the **built-in `gapminder` dataset** (`px.data.gapminder()`) to create:

## 📝 Task 9 — Choose Your Own Visualization

### Background
Good data visualization starts with choosing the **right chart type** for your data and message. This task tests that judgment.

### 🎯 Your Task
Below is a dataset of a fictional **e-commerce store's weekly performance**:

```python
data = {
    'Day':        ['Mon','Tue','Wed','Thu','Fri','Sat','Sun'],
    'Orders':     [45,   62,   58,   70,   95,   110,  80 ],
    'Revenue':    [900,  1240, 1160, 1400, 1900, 2200, 1600],
    'Returns':    [2,    3,    5,    4,    6,    8,    5   ],
    'Category':   ['Electronics','Clothing','Books','Electronics','Clothing','Books','Electronics']
}
```

**You must create exactly 3 different chart types** using any combination of Matplotlib, Seaborn, or Plotly. For each chart:
1. Clearly state (in a comment) **which chart type you chose and why** it suits that particular relationship
2. Include a title, axis labels, and any useful annotations

**Suggested angles to explore** (pick any 3):
- Trend of `Orders` or `Revenue` over the week
- Proportion of `Orders` by `Category`
- Relationship between `Orders` and `Revenue`
- Comparison of `Returns` vs `Orders` by day

> 💡 There is no single correct answer here. Focus on making each chart **clear and informative**.


````python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# dataset
data = {
    'Day': ['Mon','Tue','Wed','Thu','Fri','Sat','Sun'],
    'Orders': [45,62,58,70,95,110,80],
    'Revenue': [900,1240,1160,1400,1900,2200,1600],
    'Returns': [2,3,5,4,6,8,5],
    'Category': ['Electronics','Clothing','Books','Electronics','Clothing','Books','Electronics']
}

df = pd.DataFrame(data)

# ------------------------------
# Chart 1: Line Chart
# Reason: Line charts are best for showing trends over time.
# Here we show how Orders change during the week.
# ------------------------------

plt.figure(figsize=(8,5))
plt.plot(df['Day'], df['Orders'], marker='o')

plt.title("Trend of Orders During the Week")
plt.xlabel("Day")
plt.ylabel("Number of Orders")

plt.grid(True)
plt.show()







````

````python
# orders per category
category_orders = df.groupby('Category')['Orders'].sum()

plt.bar(category_orders.index, category_orders.values)

plt.title("Orders by Category")
plt.xlabel("Category")
plt.ylabel("Total Orders")

plt.show()
````

````python
# ------------------------------
# Chart 3: Scatter Plot
# Reason: Scatter plots show relationship between two numerical variables.
# Here we check how Orders relate to Revenue.
# ------------------------------

plt.figure(figsize=(8,5))
sns.scatterplot(x='Orders', y='Revenue', data=df)

plt.title("Relationship Between Orders and Revenue")
plt.xlabel("Orders")
plt.ylabel("Revenue")

plt.show()
````

````python
# Q1: What is the main advantage of using Seaborn over raw Matplotlib for statistical plots?
# A1: Seaborn provides a high-level interface for drawing attractive and informative statistical graphics. It simplifies the creation of complex plots, offers better default aesthetics, and integrates well with pandas DataFrames, making data exploration much more intuitive and requiring less code compared to raw Matplotlib.

# Q2: When would you choose Bokeh over Plotly, and vice versa?
# A2: Choose Bokeh for custom, highly interactive visualizations within a Python environment, especially when integrating with web applications or when a simpler, more controlled interactive experience is desired. Choose Plotly (Plotly Express) for quick, aesthetically pleasing, and highly interactive plots with built-in features like zooming, panning, and animation, often with less code. Plotly is also excellent for dashboards and web-based reporting due to its extensive ecosystem and support for various output formats.

# Q3: In Task 3, how does changing the number of bins affect your interpretation of the distribution?
# A3: Changing the number of bins significantly affects the visual interpretation of a histogram. Fewer bins can obscure details and make the distribution appear smoother or multimodal when it isn't. More bins can reveal fine details or irregularities but can also make the distribution appear noisy or sparse, especially with limited data. The ideal number of bins helps reveal the underlying shape and characteristics of the data without oversimplifying or overcomplicating it.

# Q4: Why is a pie chart generally not recommended when you have more than 5–6 categories?
# A4: Pie charts are not recommended for many categories because it becomes difficult to distinguish between the sizes of small slices and to compare the relative proportions of each category. Human eyes are not good at judging angles or areas accurately when there are many similar-sized slices, making the chart less effective for conveying information.

# Q5: What does a violin plot show that a box plot does not?
# A5: A violin plot shows the full distribution of the data (e.g., using kernel density estimation) at different levels of a categorical variable, providing a richer understanding of the data's density and shape. A box plot, in contrast, only shows summary statistics like the median, quartiles, and outliers, but does not convey the density or shape of the distribution within those quartiles.
````
